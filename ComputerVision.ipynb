```python
# show image with matplotlib
# plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) # opencv deals in bgr and matplot - rgb
plt.imshow(image) # opencv deals in bgr and matplot - rgb
plt.show()
```

```python
def imshow(title = "", image = None):
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
imshow("First Flowers", image)
```
